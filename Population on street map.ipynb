```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

urlN = ["https://northmiddle.andaman.nic.in/public-utility-category/hospitals/",
       "https://northmiddle.andaman.nic.in/public-utility-category/hospitals/page/2/"
        ]
htmldata = requests.get(urlN[0]).text
soup = BeautifulSoup(htmldata, 'html.parser')
print(soup.prettify())
```

```python
hospital = []
address = []
no=0
while no <2:
    print(no)
    url = urlN[no]
    htmldata = requests.get(url).text
    soup = BeautifulSoup(htmldata, 'html.parser')

    for el in soup.find_all('h2', {'class': 'heading4'}):
      hospital.append(el.get_text())

    masterDiv = soup.find('div', class_ = "col-12")
    for el in masterDiv.find_all('a'):
        address.append(el.get_text())
    no += 1

zipped = list(zip(hospital, address))

hdata = pd.DataFrame(zipped, columns = ['hospital', 'address'])
hdata.head()
```

```python
len(hdata)
```

```python
#south anadaman
urls = ['https://southandaman.nic.in/public-utility-category/hospitals/',
        'https://southandaman.nic.in/public-utility-category/hospitals/page/2/',
        'https://southandaman.nic.in/public-utility-category/hospitals/page/3/']

hospitalS = []
addressS = []
no = 0

while no <3:
    print(no)
    url = urls[no]
    htmldata = requests.get(url).text
    soup = BeautifulSoup(htmldata, 'html.parser')

    for el in soup.find_all('h2', {'class': 'heading4'}):
      hospitalS.append(el.get_text())

    masterDiv = soup.find('div', {'class': 'publicUtilityMainCont publicU-Details equal-height'})
    for el in masterDiv.find_all('a'):
        addressS.append(el.get_text())
    no += 1
```

```python
print(hospitalS)
```

```python
tempHosp = {'hospital': hospitalS, 'address': addressS}
hdata = pd.concat([hdata, pd.DataFrame(tempHosp)], axis = 0, ignore_index=True)
hdata
```

```python
#Sub centres have different values seperate them and make new rows by comma delimiter
# a cell into rows from 24 to 34
import numpy as np
from itertools import chain

# return list from series of comma-separated strings
def chainer(s):
    return list(chain.from_iterable(s.str.split(',')))

# calculate lengths of splits
lens = hdata1['address'].str.split(',').map(len)

# create new dataframe, repeating or chaining as appropriate
hdata3 = pd.DataFrame({'hospital': np.repeat(hdata1['hospital'], lens),
                    'address': chainer(hdata1['address'])})

hdata3 = hdata3.reset_index(drop=True)
len(hdata3)
```
